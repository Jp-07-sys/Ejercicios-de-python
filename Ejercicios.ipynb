```python
#Tengo cincuenta manzanas y quiero repartirlas en partes iguales entre 6 personas
manzanas=50
personas=6

print(manzanas//personas)

Faltan=manzanas%personas
print("Sobran",Faltan)
```

```python
#Un curso dura 100 días, cuantas semanas completas hay
curso=100
dias=7
print(curso//dias,"semanas completas")

Semanasfaltantes =curso%dias
print("Faltan",Semanasfaltantes,"semanas")
```

```python
#EJERCICIO 1
#Dado un triángulo con lados de longitudes
a=7
b=8
c=9
S=(a+b+c)/2
A=((S*(S-a)*(S-b)*(S-c)))**(1/2)
print("El semiperímetro es",S)
print("El área del triángulo es",A)
```
